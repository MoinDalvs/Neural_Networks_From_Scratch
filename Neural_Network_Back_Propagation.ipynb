### Gradient Descent

#### Learning Rate = 0.1

```python
x = np.linspace(4,6,20)
y = (x-10)**2
#plt.scatter(x,y,color='black')
starting_point=100
updated_point = starting_point
l = []
plt.figure(figsize=(12,8))
for i in range(100):
    eta=+0.1
    updated_point=updated_point-eta*2*(updated_point-10)
    l.append(updated_point)
    plt.scatter(updated_point,(updated_point-10)**2)
```

### Classification Problem uisng Neural Networks

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# laod pima indians dataset
dataset = np.loadtxt('C:/Users/Moin Dalvi/Documents/Data Science Material/Data Science/Python/Datasets/pima-indians-diabetes.data.csv',delimiter = ",")
# split into input (X) and output (Y) variables
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
dataset.shape
```

```python
dataset
```

```python
x
```

```python
y
```

```python
# create model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(12, input_dim=8,activation='sigmoid'))
model.add(tf.keras.layers.Dense(8, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(x, y, validation_split = 0.20, epochs=200, batch_size = 100)
```

```python
#model.save_weight("mywt.kmw")
```

```python
# evaluate the model
scores = model.evaluate(x, y)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1]*100))
```

```python
# Visualize training history

# list all data in history
model.history.history.keys()
import matplotlib.pyplot as plt
```

```python
history.history.keys()
```

```python
# summarize history for accuracy
plt.figure(figsize=(12,8))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.figure(figsize=(12,8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
